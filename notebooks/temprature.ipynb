temprature analysis

```python
import pandas as pd

region_1 = pd.read_pickle("no_outlier_1.pk1")

region_2 = pd.read_pickle("no_outlier_2.pk1")

region_3 = pd.read_pickle("no_outlier_3.pk1")
```

```python
region_1["Timestamp"] = pd.to_datetime(region_1['Timestamp'])
region_2["Timestamp"] = pd.to_datetime(region_2['Timestamp'])
region_3["Timestamp"] = pd.to_datetime(region_3['Timestamp'])
```

```python

import matplotlib.pyplot as plt

region_1.set_index('Timestamp', inplace=True)

# Convert 'Timestamp' index to date
region_1.index = region_1.index.date

# Scatter plot comparing Module Temperatures (TModA, TModB) with Ambient Temperature (Tamb)
plt.figure(figsize=(8, 6))

plt.scatter(region_1['TModA'], region_1['Tamb'], color='b', label='TModA vs Tamb')
plt.scatter(region_1['TModB'], region_1['Tamb'], color='r', label='TModB vs Tamb')

plt.title('Module Temperatures vs Ambient Temperature')
plt.xlabel('Module Temperature (°C)')
plt.ylabel('Ambient Temperature (°C)')
plt.legend()
plt.grid(True)

plt.show()
```
