```python
import pandas as pd
region_1 = pd.read_pickle("region_1.pk1")
region_2 = pd.read_pickle("region_2.pk1")
region_3 = pd.read_pickle("region_3.pk1")
```

now lets check for any empty values

```python
import missingno as msno

msno.bar(region_1)
```

```python
msno.bar(region_2)
```

```python
msno.bar(region_3)
```

as we can see from the above, all the comments are empty. so we can delete Comments column.

```python
no_comments_1 = region_1.drop(columns=["Comments"])
no_comments_2 = region_2.drop(columns=["Comments"])
no_comments_3 = region_3.drop(columns=["Comments"])

no_comments_1.to_pickle("no_comments_1.pk1")
no_comments_2.to_pickle("no_comments_2.pk1")
no_comments_3.to_pickle("no_comments_3.pk1")
```
