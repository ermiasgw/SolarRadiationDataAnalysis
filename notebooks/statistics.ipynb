```python
import pandas as pd

region_1 = pd.read_pickle("no_comments_1.pk1")

region_2 = pd.read_pickle("no_comments_2.pk1")

region_3 = pd.read_pickle("no_comments_3.pk1")
```

lets check for mean, varience ... for these datas

```python
region_1.describe()
```

```python
region_2.describe()
```

```python
region_3.describe()
```

as we can see from the calculations there are negative values in GHI, DNI, DHI. those were supposed to be positive numbers.
so our move is to remove all rows which contain negative numbers.

```python
no_negative_1 = region_1.drop(region_1[(region_1.GHI < 0) | (region_1.DNI < 0) | (region_1.DHI <0)].index)

no_negative_2 = region_2.drop(region_2[(region_2.GHI < 0) | (region_2.DNI < 0) | (region_2.DHI <0)].index)

no_negative_3 = region_3.drop(region_3[(region_3.GHI < 0) | (region_3.DNI < 0) | (region_3.DHI <0)].index)
```

now lets check it.

```python
no_negative_1.describe()
```

lets save the data.

```python
no_negative_1.to_pickle("no_negative_1.pk1")
no_negative_2.to_pickle("no_negative_2.pk1")
no_negative_3.to_pickle("no_negative_3.pk1")
```
