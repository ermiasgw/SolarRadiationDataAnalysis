lets analyze the time series

```python
import pandas as pd

region_1 = pd.read_pickle("no_outlier_1.pk1")

region_2 = pd.read_pickle("no_outlier_2.pk1")

region_3 = pd.read_pickle("no_outlier_3.pk1")
```

for region 1 and convert timestamp from object to datetime

```python
region_1["Timestamp"] = pd.to_datetime(region_1['Timestamp'])
region_2["Timestamp"] = pd.to_datetime(region_2['Timestamp'])
region_3["Timestamp"] = pd.to_datetime(region_3['Timestamp'])
region_1.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


region_1.replace([np.inf, -np.inf], np.nan, inplace=True)

region_1_hourly = region_1.set_index('Timestamp')

# Resample the DataFrame on an hourly basis and calculate the mean
region_1_hourly = region_1_hourly.resample('D').mean()

# Plotting the metrics
plt.figure(figsize=(15, 10))

# Plot GHI
plt.subplot(4, 1, 1)
sns.lineplot(data=region_1_hourly['GHI'], marker='o', color='blue')
plt.title('Hourly GHI')
plt.xlabel('Timestamp')
plt.ylabel('GHI')

# Plot DNI
plt.subplot(4, 1, 2)
sns.lineplot(data=region_1_hourly['DNI'], marker='o', color='green')
plt.title('Hourly DNI')
plt.xlabel('Timestamp')
plt.ylabel('DNI')


# Plot DHI
plt.subplot(4, 1, 3)
sns.lineplot(data=region_1_hourly['DHI'], marker='o', color='red')
plt.title('Hourly DHI')
plt.xlabel('Timestamp')
plt.ylabel('DHI')

# Plot Tamb
plt.subplot(4, 1, 4)
sns.lineplot(data=region_1_hourly['Tamb'], marker='o', color='orange')
plt.title('Hourly Tamb')
plt.xlabel('Timestamp')
plt.ylabel('Tamb')


plt.tight_layout()
plt.show()
```
