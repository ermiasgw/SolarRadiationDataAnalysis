scatter plots

```python
import pandas as pd

region_1 = pd.read_pickle("no_outlier_1.pk1")

region_2 = pd.read_pickle("no_outlier_2.pk1")

region_3 = pd.read_pickle("no_outlier_3.pk1")
```

for region 1

```python
import matplotlib.pyplot as plt


fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Scatter plot for GHI vs. Tamb
axs[0].scatter(region_1['GHI'], region_1['Tamb'], color='b')
axs[0].set_title('GHI vs. Tamb')
axs[0].set_xlabel('GHI')
axs[0].set_ylabel('Tamb')
axs[0].grid(True)

# Scatter plot for WS vs. WSgust
axs[1].scatter(region_1['WS'], region_1['WSgust'], color='r')
axs[1].set_title('WS vs. WSgust')
axs[1].set_xlabel('WS')
axs[1].set_ylabel('WSgust')
axs[1].grid(True)

# Scatter plot for DNI vs. DHI
axs[2].scatter(region_1['DNI'], region_1['DHI'], color='g')
axs[2].set_title('DNI vs. DHI')
axs[2].set_xlabel('DNI')
axs[2].set_ylabel('DHI')
axs[2].grid(True)

# Adjust layout
plt.tight_layout()

# Show scatter plots
plt.show()
```
