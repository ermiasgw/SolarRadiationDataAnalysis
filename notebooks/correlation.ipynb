correlation analysis

```python
import pandas as pd

region_1 = pd.read_pickle("no_outlier_1.pk1")

region_2 = pd.read_pickle("no_outlier_2.pk1")

region_3 = pd.read_pickle("no_outlier_3.pk1")
```

for region 1

```python
import seaborn as sns
import matplotlib.pyplot as plt

selected_columns = ['GHI', 'DNI', 'DHI', 'TModA', 'TModB']
df_selected_1 = region_1[selected_columns]

correlation_matrix = df_selected_1.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=.5)
plt.title('Correlation Matrix (GHI, DHI, DNI vs TModA)')
plt.show()
```
